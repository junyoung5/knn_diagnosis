```r
#Knn Diagnosing Breast Cancer

library(class)
wbcd <- read.csv("https://github.com/hbchoi/SampleData/raw/master/wisc_bc_data.csv", stringsAsFactors = F)
str(wbcd)
wbcd <- wbcd[,-1]
table(wbcd$diagnosis)
wbcd$diagnosis <- ifelse(wbcd$diagnosis == 'B', 'Benign', 'Malignant')


#Normalization: make all variables to have same scale
minmax_norm <- function(x){
  (x-min(x))/(max(x)-min(x))
}

wbcd_norm <- sapply(wbcd[,-1], minmax_norm)

#split data
dim(wbcd_norm)

wbcd_train <- wbcd_norm[1:469, ] 
wbcd_test <- wbcd_norm[470:569, ]
wbcd_train_label <- wbcd[1:469, 1] 
wbcd_test_label <- wbcd[470:569, 1]

sqrt(nrow(wbcd_train))
wbcd_test_pred <- knn(train = wbcd_train, test = wbcd_test, cl = wbcd_train_label, k = 21)

#Performance
cmat <- table(wbcd_test_pred, wbcd_test_label)
cmat
accuracy <- (cmat[1,1] + cmat[2,2]) / sum(cmat)
precision <- (cmat[2,2])/ (cmat[2,2] + cmat[2,1])
recall <- cmat[2,2] / (cmat[2,2] + cmat[1,2])
#recall: 0.948   â€”> we want to avoid false negative how can? 


#probabilistic interpretation
wbcd_test_pred <- knn(train = wbcd_train, test = wbcd_test, cl = wbcd_train_label, k = 21, prob = TRUE)
head(wbcd_test_pred)
attributes(wbcd_test_pred)
head(attributes(wbcd_test_pred)$prob)

#converting all prob to Malignant
wbcd_test_pred_prob <- ifelse(wbcd_test_pred == 'Malignant', attributes(wbcd_test_pred)$prob, 1-attributes(wbcd_test_pred)$prob)
head(wbcd_test_pred_prob)
summary(wbcd_test_pred_prob)

library(ROCR)
plot(performance(prediction(wbcd_test_pred_prob, wbcd_test_label == 'Malignant'),
                 'tpr', 'fpr'))

calAUC <- function(predCol, targetCol){
  perf <- performance(prediction(predCol, targetCol), 'auc')
  as.numeric(perf@y.values)
}

calAUC(wbcd_test_pred_prob, wbcd_test_label == 'Malignant')

#Adjusting Threshold
threshold <- 0.1
wbcd_test_pred_new <- ifelse(wbcd_test_pred_prob > threshold, 'Malignant', 'Benighn')
cmat <- table(wbcd_test_pred_new, wbcd_test_label)
cmat
(cmat[1,1]+cmat[2,2]) / sum(cmat)
cmat[2,2] / (cmat[2,1] + cmat[2,2])
cmat[2,2] / sum(cmat[,2])
```
